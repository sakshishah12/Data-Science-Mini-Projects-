```python
import pandas as pd
import numpy as np
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies_df=pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies_df.head()
```

```python
print("Credits:",credits.shape)
print("Movies:",movies_df.shape)
```

```python
credits_column_renamed=credits.rename(index=str,columns={"movie_id":"id"})
movies_df_merge=movies_df.merge(credits_column_renamed,on="id")
movies_df_merge.head()
```

```python
movies_cleaned_df=movies_df_merge.drop(columns=["title_x","title_y","homepage","status","production_companies"])
movies_cleaned_df.head()
```

```python
movies_cleaned_df.info()
```

```python
v = movies_cleaned_df["vote_count"]
R=movies_cleaned_df["vote_average"]
C=movies_cleaned_df["vote_average"].mean()
m=movies_cleaned_df["vote_count"].quantile(0.70)
```

```python
movies_cleaned_df["weighted_avg"]=((R*v) + (C*m))/(v+m)
```

```python
movies_cleaned_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies_sorted_ranking=movies_cleaned_df.sort_values('weighted_avg',ascending=False)
movies_sorted_ranking[['original_title','vote_count','vote_average','weighted_avg','popularity']]

plt.figure(figsize=(10,6))
ax=sns.barplot(x=movies_sorted_ranking['weighted_avg'].head(10),y=movies_sorted_ranking['original_title'].head(10),data=movies_sorted_ranking)
plt.title("Best by average votes")
plt.xlabel('Score of weighted average')
plt.ylabel('Movie Title')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
popularity=movies_cleaned_df.sort_values('popularity',ascending=False)
plt.figure(figsize=(10,6))
ax=sns.barplot(x=popularity['popularity'].head(10),y=popularity['original_title'].head(10),data=popularity)
plt.title("Most Popular by votes")
plt.xlabel('Score of Popularity')
plt.ylabel('Movie Title')
```

```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(movies_cleaned_df[['weighted_avg','popularity']])
movie_normalized_df=pd.DataFrame(movie_scaled_df,columns=['weighted_avg','popularity'])
movie_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_weight_average','normalized_popularity']]= movie_normalized_df
```

```python
movies_cleaned_df['score'] = movies_cleaned_df['normalized_weight_average'] * 0.5 + movies_cleaned_df['normalized_popularity'] * 0.5
movies_scored_df = movies_cleaned_df.sort_values(['score'], ascending=False)
movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```
